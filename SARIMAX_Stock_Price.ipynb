```python

import statsmodels.api as sm
import numpy as np
np.random.seed(100)

data = np.sort(np.random.uniform(0, 1, size=30))
steps_to_predict = 5

model = sm.tsa.statespace.SARIMAX(endog=data,order=(2,0,0),enforce_stationarity=False)
sarima = model.fit()
print(sarima.summary())
```
